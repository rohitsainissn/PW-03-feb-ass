Q1. What is the mathematical formula for a linear SVM?

Ans:- The mathematical formula for a linear SVM (Support Vector Machine) is defined as follows:

Given a training set of n labeled examples (x1, y1), (x2, y2), ..., (xn, yn), where xi represents the feature vector of the ith example, and yi is the class label (+1 or -1), the objective of a linear SVM is to find a hyperplane in the feature space that maximizes the margin between the two classes.

The hyperplane can be represented by the equation:

w^T x + b = 0

where w is the weight vector perpendicular to the hyperplane, x is the feature vector of an example, and b is the bias term.

The margin between the hyperplane and the closest positive and negative examples (known as support vectors) is given by:

2 / ||w||

where ||w|| is the Euclidean norm of the weight vector.

The optimal hyperplane is found by solving the following optimization problem:

minimize: 1/2 ||w||^2

subject to: yi (w^T xi + b) >= 1

where ||w||^2 is the square of the Euclidean norm of the weight vector, and yi is the class label of the ith example.

The optimization problem can be solved using a quadratic programming algorithm, which finds the values of w and b that minimize the objective function while satisfying the constraints. The resulting hyperplane can then be used to classify new examples based on which side of the hyperplane they fall on.

Hyperplane:
A hyperplane is a decision boundary that separates two classes in a feature space. It is defined by the equation:
w^T x + b = 0

where w is the weight vector perpendicular to the hyperplane, x is the feature vector of an example, and b is the bias term.


Marginal plane:
A marginal plane is a decision boundary that separates two classes with a margin that is greater than zero. It is defined by the equation:
w^T x + b = +/-1

where w is the weight vector perpendicular to the hyperplane, x is the feature vector of an example, b is the bias term, and +/-1 is the margin.

Hard margin:
A hard margin is a decision boundary that does not allow for any misclassifications. It is used when the classes are linearly separable.
Here is an example of a hard margin SVM that separates two classes:

In this example, the two classes are represented by the red and blue dots, and the hard margin SVM is represented by the solid black line. The hard margin SVM does not allow for any misclassifications and separates the two classes with a clear boundary.


Soft margin:
A soft margin is a decision boundary that allows for some misclassifications. It is used when the classes are not linearly separable. Soft margin SVM adds a penalty term to the objective function that allows for some misclassifications.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.svm import SVC

# Load iris dataset
iris = load_iris()

# Extract two features for visualization
X = iris.data[:, :2]
y = iris.target

# Create an SVM classifier with a linear kernel
svm_clf = SVC(kernel='linear', C=1)
svm_clf.fit(X, y)

# Define the grid of points to plot the decision boundary
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))

# Predict the labels for the grid points
Z = svm_clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary and the data points
plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('SVM Decision Boundary')
plt.show()
```
