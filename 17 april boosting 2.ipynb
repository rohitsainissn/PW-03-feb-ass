Q1. What is Gradient Boosting Regression?

Ans:- Gradient boosting regression is a boosting technique where we first make the base model which is average of outcome, then residuals are calculated, the residuals are error from the actual value, then we train our model with dependent variable as x and residual as y.on basis of our prediction residual gets updated and same step repeat.

Final output is = base model output + sum of (learning rate * output of models)


Q2. Implement a simple gradient boosting algorithm from scratch using Python and NumPy. Use a
simple regression problem as an example and train the model on a small dataset. Evaluate the model's
performance using metrics such as mean squared error and R-squared.


```python
df= sns.load_dataset("tips")
df
```

```python
df = pd.get_dummies(df, columns=['sex','smoker','day', 'time'], drop_first=True)
df
```

```python
x = df.drop(['tip'],axis=1)
y = df[['tip']]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test  = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor(random_state=0)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

Q3. Experiment with different hyperparameters such as learning rate, number of trees, and tree depth to
optimise the performance of the model. Use grid search or random search to find the best
hyperparameters

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for GridSearchCV
param_grid = {'n_estimators': [50, 100, 200],
              'loss':['squared_error', 'absolute_error', 'huber', 'quantile'],
              'max_depth': [2, 3, 4],
              'learning_rate': [0.1, 0.01, 0.001]}

# Create a GradientBoostingRegressor object
model = GradientBoostingRegressor(random_state=0)

# Create a GridSearchCV object
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=5)

# Fit the GridSearchCV object on the training data
grid.fit(X_train, y_train)

# Print the best parameter values and the corresponding mean score
print(f"Best parameters: {grid.best_params_}")
print(f"Best score: {grid.best_score_:.2f}")
```

Q4. What is a weak learner in Gradient Boosting?

Ans: week learner is the decision trees with small depth like 3, these small trees gives output by learning independent variable and residual, then we predict the result using old value + learning rate * residual
then we calculate new residual from th prediction and obtain new prediction.


These week learners reduces the overfitting, hence reduce variance, as many models are used then bias is also reduced.
