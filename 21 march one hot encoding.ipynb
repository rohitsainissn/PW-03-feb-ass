Q4. For a dataset with the following categorical variables: Color (red, green, blue), Size (small, medium, large), and Material (wood, metal, plastic), perform label encoding using Python's scikit-learn library. Show your code and explain the output.

Ans for Color (red, green, blue) and Material (wood, metal, plastic) we use one hot encoding and for Size (small, medium, large) : Ordinal Encoding, 



```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
import pandas as pd


data = {'Color': ['red', 'green', 'blue', 'red', 'blue'], 
        'Size': ['small', 'medium', 'large', 'medium', 'small'], 
        'Material': ['wood', 'metal', 'plastic', 'wood', 'plastic']}
df = pd.DataFrame(data)


ordinal = OrdinalEncoder()
onehot = OneHotEncoder()


df['Size'] = ordinal.fit_transform(df[['Size']])


onehot_encoded = onehot.fit_transform(df[['Color', 'Material']]).toarray()
onehot_df = pd.DataFrame(onehot_encoded, columns=onehot.get_feature_names_out(['Color', 'Material']))

# concatenate original and encoded DataFrames
df_encoded = pd.concat([df.drop(['Color', 'Material'], axis=1), onehot_df], axis=1)

print(df_encoded)
```
