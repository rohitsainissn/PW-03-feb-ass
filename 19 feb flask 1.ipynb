Q1. What is Flask Framework? What are the advantages of Flask Framework?

Ans:- Flask is a lightweight and flexible Python web framework that allows you to build web applications quickly and easily. It is classified as a "micro-framework" because it does not require particular tools or libraries and has no database abstraction layer or form validation features by default, which gives you the flexibility to use other libraries and tools that you prefer. Flask provides only the essentials for web development, such as routing, templating, and request/response handling.
