In linear regression we will make the best fit line and choose a threshold below which we said student will fail, while in logistic regression model, the curve will overlap most of the point and below the threshold all student are fail.

Q4. What is the ROC curve, and how is it used to evaluate the performance of the logistic regression
model?

Ans:- ROC curve is the plot between true positive rate and false positive rate. the area under the curve tells the performance of the model. This graph is used to find the threshold for the model whether we need good recall or good precision.


Q5. What are some common techniques for feature selection in logistic regression? How do these
techniques help improve the model's performance?

Ans:- Forward feature sellection, backward feature sellection, principle component analysis and l1 regularisation are some of the feature selection technique used in logistic regression. these technique removes the unimportant features from the data set hence complexicity of the model get reduced and model can train itself with important features hence performance improves.

Q6. How can you handle imbalanced datasets in logistic regression? What are some strategies for dealing
with class imbalance?

ANs:- Imbalance data set can be handled by :
1. making synthetic data points
2. increasing weight of minority data points.
3. removing majority data points if data set is very large.

Q7. Can you discuss some common issues and challenges that may arise when implementing logistic
regression, and how they can be addressed? For example, what can be done if there is multicollinearity
among the independent variables?

Ans:- Overfitting, imbalance data set, outliers, missing values are common issues that arrises when implementing logistic regression. We can overfitting by using l1 and l2 regularisation term, imbalance data set can be made balanced by using synthetic data points, increase weight of minority class, or removing majority class extra point.


Multicollinearity: Multicollinearity occurs when two or more independent variables in the model are highly correlated. This can lead to unreliable or unstable estimates of the model coefficients. To address multicollinearity, one can drop one of the correlated variables, combine them to create a new variable, or use techniques such as principal component analysis (PCA) to reduce the number of variables.
