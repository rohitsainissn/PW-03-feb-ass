Q1. What is the role of feature selection in anomaly detection?

Feature selection plays a crucial role in anomaly detection by determining which features or variables are most relevant for identifying anomalies in a dataset. It involves choosing a subset of features from the original set of variables to improve the effectiveness and efficiency of anomaly detection algorithms. By selecting the most informative and discriminative features, feature selection helps to reduce dimensionality, eliminate irrelevant or redundant features, and enhance anomaly detection performance.

Q2. What are some common evaluation metrics for anomaly detection algorithms and how are they computed?

True Positive (TP): The number of true anomalies correctly identified by the algorithm.

False Positive (FP): The number of non-anomalies incorrectly identified as anomalies by the algorithm.

True Negative (TN): The number of non-anomalies correctly identified as non-anomalies by the algorithm.

False Negative (FN): The number of true anomalies missed or not identified by the algorithm.

Using these metrics, several evaluation measures can be computed:

True Positive Rate (TPR) or Recall: TPR = TP / (TP + FN), measures the proportion of actual anomalies correctly detected.

False Positive Rate (FPR): FPR = FP / (FP + TN), measures the proportion of non-anomalies incorrectly classified as anomalies.

Precision: Precision = TP / (TP + FP), measures the proportion of correctly identified anomalies out of all instances classified as anomalies.

F1 Score: F1 Score = 2 * (Precision * Recall) / (Precision + Recall), harmonic mean of precision and recall.

Area Under the Receiver Operating Characteristic Curve (AUC-ROC): Computes the area under the ROC curve, which plots TPR against FPR at various threshold settings.

Average Precision (AP): Computes the average precision at various recall levels, providing a summary measure of the precision-recall trade-off.

F-beta Score: F-beta Score = (1 + beta^2) * (Precision * Recall) / ((beta^2 * Precision) + Recall), a generalized version of the F1 score that allows adjusting the trade-off between precision and recall using the parameter beta.

Q3. What is DBSCAN and how does it work for clustering?

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a density-based clustering algorithm that groups together data points that are close to each other in feature space and separates them from areas of lower density. It does not require specifying the number of clusters in advance and is effective in identifying clusters of arbitrary shape.

DBSCAN works by defining a neighborhood around each data point based on a distance threshold (epsilon) and a minimum number of points (minPts). It categorizes data points into three types: core points, border points, and noise points.

Q8. What are local outliers and global outliers, and how do they differ from each other?

Local Outliers: Local outliers, also known as contextual outliers or micro outliers, are data points that are considered unusual or abnormal within a specific neighborhood or local region of the dataset. They are identified by comparing the characteristics or behavior of a data point to its neighboring points. A data point may be a local outlier in one region but not in another.

Global Outliers: Global outliers, also referred to as collective outliers or macro outliers, are data points that exhibit anomalous behavior when considering the entire dataset as a whole. These outliers are identified by examining the overall distribution or patterns in the data, disregarding specific local regions. Global outliers may have characteristics that make them stand out from the majority of the data points.

Q9. How can local outliers be detected using the Local Outlier Factor (LOF) algorithm?

Compute the distance between each data point and its neighboring points.
Determine the reachability distance for each point, which represents the minimum distance required to reach a point from its neighbors.
Calculate the local reachability density for each point, which is the inverse of the average reachability distance of its neighbors.
Compute the Local Outlier Factor (LOF) for each point by comparing its local reachability density to that of its neighbors. LOF measures how isolated a point is compared to its neighbors.
Points with a significantly higher LOF value compared to their neighbors are considered local outliers.

Q11. What are some real-world applications where local outlier detection is more appropriate than global outlier detection, and vice versa?

Local Outlier Detection:

Anomaly detection in sensor networks: Local outlier detection is suitable for identifying anomalies or malfunctions at a specific sensor location without considering the global behavior of the entire network.

Fraud detection in credit card transactions: Local outlier detection can detect unusual spending patterns or transactions for individual credit card holders, allowing for targeted fraud detection at the customer level.

Intrusion detection in computer networks: Analyzing local network traffic patterns can help detect unusual activities or intrusions at specific hosts or subnetworks rather than relying on global network behavior.


Global Outlier Detection:

Financial market analysis: Global outlier detection is valuable for identifying major market events that affect the entire financial system, such as market crashes or significant price fluctuations across multiple stocks or assets.

Disease outbreak detection: Global outlier detection can identify regions or populations experiencing a significantly higher incidence of a disease, indicating potential outbreaks that require attention and intervention.

Quality control in manufacturing: Global outlier detection helps identify systemic issues in the manufacturing process by detecting anomalies in the overall distribution of product measurements or quality attributes.
